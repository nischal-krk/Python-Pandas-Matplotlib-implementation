```python
import pandas as pd                                     #import pandas
import matplotlib.pyplot as plt                         #import pyplot from matplotlib
```

```python
dat = pd.read_csv("sales_record.csv")                   #assign the csv file to a variable
```

```python
dat.columns                                             #display the headers
```

```python
dat.iloc[1:4]                                          #display data from index location specification
```

```python
dat['Country'][1:40]                                   #display certain column from index specification      
```

```python
dat["Total Cost"].mean()                     #statistical summary for mean
```

```python
dat["Total Profit"].median()                 #statistical summary for median
```

```python
dat["Total Profit"].min()                    #statistical summary for minimum value
```

```python
dat["Total Profit"].max()                     #statistical summary for maximum vlaue
```

```python
xval = dat["Total Cost"].tolist               #converting pandas dataframe to a list using .tolist() method 
yval = dat["Total Profit"].tolist()
```

```python
plt.plot(xval,yval)                           #drawing a line plot
plt.xlabel("Total Cost")                      #specifying the x-label
plt.ylabel("Total Profit")                    #specifying the y-label
```

```python
plt.scatter(xval,yval)                       #drawing a scatter plot 
plt.xlabel("Total Cost")
plt.ylabel("Total Profit")
```

```python
plt.hist(xval,bins=15)                           #drawing a histogram with 15 bins
plt.xlabel("Total Cost")
```

```python
plt.hist(yval,bins=40)                           #drawing a histogram with 40 bins
plt.xlabel("Total Profit")
```
